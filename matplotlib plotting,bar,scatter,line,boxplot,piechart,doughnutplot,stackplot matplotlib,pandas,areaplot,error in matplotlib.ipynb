# Matplotlib Python Plotting Ways (Types)


There are various plotting techniques or ways that can be carried out on the data provided, and some of these plotting types are as follows:

# Line Plot

The plotting of the frequency of data along a line can be represented using a line plot. It is one of the simplest and commonly used plotting methods. Line plotting is a primitive plotting technique as it is a plotting method that was first introduced.

```python
plt.plot(x,y,label='x-y',linewidth=10)
plt.plot(x,y2,label='x-y2',linewidth=10)
plt.title('First REPRESENTATION') #table name
plt.xlabel('X-axis')   #x-axis Name
plt.ylabel('Y-axis')   #y-axis Name
plt.grid()    #use for grid in graph
plt.legend()
```

# Bar plot

Categorical data can be represented in rectangular blocks with different heights or lengths proportional to the values. Such a type of representation is called a bar chart. Bar charts can be used to plot data in both vertical and horizontal manner.

```python
plt.figure(1)
plt.bar(x,y2,label='x-y2',color='orange')
plt.bar(x,y,label='x-y')
plt.title('Bar chart REPRESENTATION') #table name
plt.xlabel('X-axis')   #x-axis Name
plt.ylabel('Y-axis')   #y-axis Name
plt.grid()    #use for grid in graph
plt.legend()



plt.figure(2)
plt.bar(x,y2,label='x-y2',color='orange')
plt.title('Bar chart REPRESENTATION') #table name
plt.xlabel('X-axis')   #x-axis Name
plt.ylabel('Y-axis')   #y-axis Name
plt.grid()    #use for grid in graph
plt.legend()


plt.figure(3)
plt.barh(x,y2,label='x-y2',color="maroon")
plt.barh(x,y,label='x-y',color='orange')
plt.title('Bar chart REPRESENTATION') #table name
plt.xlabel('X-axis')   #x-axis Name
plt.ylabel('Y-axis')   #y-axis Name
plt.grid()    #use for grid in graph
plt.legend()
```

# Area Plot

This type of plotting is basically used for quantitative data. A line chart forms the basis of an area plot, where the region between the axis and the line is represented by colors.

```python
#area plot is not more use data science
plt.stackplot(x,y2)
plt.stackplot(x,y)

plt.legend(["line1","line2"],loc="best")
```

# Scatter Plot

Dot-based plotting of multiple variables along x and y axes represents a scatter plot.

```python
#y3=[3,2,6,4,8,9,0,4,2]
plt.scatter(x,y,color='orange',edgecolor='blue',s=200,alpha=0.9,marker='o')
plt.scatter(x,y2,s=200,color='blue',marker='4')
plt.title('Scatter REPRESENTATION') #table name
plt.xlabel('X-axis')   #x-axis Name
plt.ylabel('Y-axis')   #y-axis Name
plt.grid()    #use for grid in graph
plt.legend(['x-y','x-y2'],loc='best')
```

# Histogram Plot

The plotting of numerical data in a precise manner by using rectangular blocks forms the basis of histogram plotting. A probability distribution can be estimated using a histogram plot. The data is mostly represented in a continuous manner based on the data set provided to plot the graph.

```python
plt.hist(x ,bins=10)
```

# Pie Plot

In a pie plot, statistical data can be represented in a circular graph where the circle is divided into portions that denote a particular data, that is, each portion, called a slice, is proportional to different values in the data. This sort of a plot can be mainly used in mass media and business.

```python
label=['bjp','cong','sp','bsp','other']
sizes=[312,7,47,19,18]

plt.pie(sizes,labels=label,autopct='%1.1f%%',startangle=90,shadow=True,explode=(1,0,0,0,0))
```

# doughnut chart

doughnut is a simular to pie chart 
#its a combination of two pia chart

```python
label=['bjp','cong','sp','bsp','other']
sizes=[312,7,47,19,18]
size_center=[5]
plt.pie(sizes,labels=label,radius=1.5,startangle=90,shadow=True,explode=(0,0,0,0,0))
plt.pie(size_center,radius=1.0,colors="w")
plt.show()
```

# Boxploting

This graph represents the minimum, maximum, median, first quartile and third quartile in the data set.

```python
available=[20,40,50,28,70]
sell=[12,20,30,23,35]
discount=[10,20,40,60,20]
data=list([available,sell,discount])
#data
plt.boxplot(data,patch_artist=True,labels=['available','sell','discount'])
#notch=True creates the notch format of the box plot
#patch_artist=True, fills the boxplot with color
```

```python
plt.violinplot(data)
```
